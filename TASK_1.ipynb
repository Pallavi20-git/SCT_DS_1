```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv', skiprows=4)
df.head()
```

```python
df_2020 = df[['Country Name', '2020']].dropna()
df_2020.columns = ['Country', 'Population']
df_2020 = df_2020.sort_values(by='Population', ascending=False)
df_2020.head()
```

```python
top10 = df_2020.head(10)

plt.figure(figsize=(12,6))
sns.barplot(data=top10, x='Country', y='Population')
plt.xticks(rotation=45)
plt.title('Top 10 Most Populous Countries in 2020')
plt.xlabel('Country')
plt.ylabel('Population')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df_2020['Population'], bins=30, kde=True)
plt.title('Histogram: Population Distribution in 2020')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```
